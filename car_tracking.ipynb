### Reading and writing the new video

```python
os.chdir('D:\WorkStation\object\computer_vision\car_traking')
cap =  cv.VideoCapture('highway.mp4')
fourcc = cv.VideoWriter_fourcc('X','V','I','D')
out = cv.VideoWriter('out.avi',fourcc,20.0,(1280,720))
```

### Ecludin algorithm for tracking and countours for object detection

```python
track = EuclideanDistTracker()
obj_detection = cv.createBackgroundSubtractorMOG2(history=100,varThreshold=50)
while(True):
    ret,frame = cap.read()
    if ret is not True:
        break
    height, width, _ = frame.shape
    ROI = frame[340: 720,500: 800]
    mask = obj_detection.apply(ROI)
    _,mask = cv.threshold(mask,254,255,cv.THRESH_BINARY)
    contours,_ = cv.findContours(mask,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
    detection = []
    for cnt in contours:
        if cv.contourArea(cnt)>100:
            (x,y,w,h) = cv.boundingRect(cnt)
            detection.append([x,y,w,h])
    boxes = track.update(detection)
    for box in boxes:
        (x,y,w,h,id) = box
        cv.putText(ROI,str(id),(x,y),cv.FONT_HERSHEY_PLAIN,2,(255,255,0),2)
        cv.rectangle(ROI,(x,y),(x+w,y+h),(0,255,0),3)
    cv.imshow('ROI',ROI)
    cv.imshow('Frame',frame)
    cv.imshow('mask',mask)
    frame = cv.resize(frame, (1280,720))
    out.write(frame)
    k = cv.waitKey(30)
    if k==27:
        break
out.release()
cap.release()
cv.destroyAllWindows()
    
```
